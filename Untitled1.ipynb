```python
votes = pd.read_csv('https://raw.githubusercontent.com/tonmcg/US_County_Level_Election_Results_08-16/master/2016_US_County_Level_Presidential_Results.csv')
#Remove " County" for clean merge
votes['county_name'] = votes['county_name'].str.replace(' County', '')
print(votes.shape)
print(votes['county_name'].value_counts().nlargest(25))
votes.head()
```

```python
path_to_shape = os.path.join('UScounties', 'UScounties.shp')
shapefile = gpd.read_file(path_to_shape)
shapefile = shapefile.rename(columns={'NAME':'county_name'})
print(shapefile.shape)
print(shapefile['county_name'].value_counts().nlargest(10))
shapefile.head()
```

```python
# shapefile = shapefile[['county_name', 'STATE_NAME', 'geometry']].merge(votes,
#                                                            on='county_name', how='left')
votes = shapefile.merge(votes, on='county_name', how='left')
print(votes['county_name'].value_counts().nlargest(25))
print(votes.shape)
votes.head()
# shapefile.drop(['total_votes', 'diff', 'votes_dem', 'votes_gop'], axis=1, inplace=True)
```
