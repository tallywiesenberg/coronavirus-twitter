# Imports

```python
from model import Tweets, engine
from streamer import session

import os

import eli5
import en_core_web_sm
from gensim.utils import simple_preprocess
from gensim.parsing.preprocessing import preprocess_documents, preprocess_string
import geopandas as gpd
from geopandas import GeoDataFrame
import numpy as np
import pandas as pd
from sklearn.feature_extraction.text import CountVectorizer
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score, roc_auc_score
from sklearn.model_selection import train_test_split
from sklearn.naive_bayes import MultinomialNB, GaussianNB
from sklearn.pipeline import Pipeline
import spacy

nlp = en_core_web_sm.load()
```

# Loading Data

## From Tweets Database

```python
df = pd.read_sql_table("tweets", con=engine)

#Convert to GeoDataFrame with coordinates
gdf = GeoDataFrame(
    df,
    geometry=gpd.points_from_xy(df['longitude'], df['latitude']))
print(gdf.shape)
gdf.head()
```

## From 2016 Presidential Election Results CSV and US States CSV

```python
votes = pd.read_csv('https://raw.githubusercontent.com/tonmcg/US_County_Level_Election_Results_08-16/master/2016_US_County_Level_Presidential_Results.csv')
path_to_states = os.path.join('data', 'states.csv')
states = pd.read_csv(path_to_states)

#Access both states and state abbreviations
votes = votes.merge(states['State'], how='left', left_on='state_abbr', right_on=states['Code'])
#Rename state column for merge with shapefile
votes = votes.rename(columns={'State':'STATE_NAME'})
#Remove " County" for clean merge with shapefile
votes['county_name'] = votes['county_name'].str.replace(' County', '')
print(votes.shape)
votes.head()
```

## From US Counties Shapefile

```python
path_to_shape = os.path.join('data', 'UScounties', 'UScounties.shp')
shapefile = gpd.read_file(path_to_shape)
shapefile = shapefile.rename(columns={'NAME':'county_name'})
print(shapefile.shape)
shapefile.head()
```

```python
votes = shapefile.merge(votes, on=['county_name', 'STATE_NAME'], how='inner')
print(votes.shape)
votes.head()
```

## Penultimate DataFrame: Counties, Geometries, and Election Results

```python
#Drop unwanted columns

votes = votes.drop(['votes_dem', 'votes_gop', 'Unnamed: 0',
                    'total_votes', 'diff', 'state_abbr',
                    'combined_fips', 'STATE_FIPS', 'CNTY_FIPS',
                    'FIPS'], axis=1)
#New columns: Blue or Red
votes['winner'] = ['Blue' if votes['per_dem'][i] > votes['per_gop'][i] else 'Red' for i in range(votes.shape[0])]
votes.head()
```

## Geopandas: Match tweet to county

```python
def match_to_county(tweet):
    for i in range(votes.shape[0]):
        if votes['geometry'][i].contains(tweet['geometry']):
            return votes['county_name'][i]

def match_to_state(tweet):
    for i in range(votes.shape[0]):
        if votes['geometry'][i].contains(tweet['geometry']):
            return votes['STATE_NAME'][i]
```

```python
match_to_county(gdf.iloc[20, :])
```

```python
gdf['county_name'] = [match_to_county(gdf.iloc[i, :]) for i in range(gdf.shape[0])]
```

```python
gdf['STATE_NAME'] = [match_to_state(gdf.iloc[i, :]) for i in range(gdf.shape[0])]
```

## Final merge

```python
final = pd.DataFrame(gdf.merge(votes[['county_name', 'STATE_NAME', 'per_point_diff', 'winner']], on=['county_name', 'STATE_NAME']))

#Convert timestamp column to datetime
final['timestamp'] = pd.to_datetime(final['timestamp'])

#Convert tweet Series to dtype String
final['tweet'] = final['tweet'].astype('str')

print(final.shape)
final.head()
```

```python
final['timestamp'].dtype
```

# Models

### Custom tokenizer

```python
def tokenize(doc):
    return [token.text for token in nlp(doc) 
            if (token.pos_ not in set(['PROPN', 'PUNCT', 'ADP', 'NUM', 'SPACE', 'PRON', 'DET', 'AUX', 'PART']))
            and (token.text not in set(['#', 'coronavirus', 'covid', '𝚌𝚘𝚛𝚘𝚗𝚊𝚟𝚒𝚛𝚞𝚜']))
            and token.like_url == False
            and token.is_stop == False
           ]
```

```python
tokenize(final['tweet'][70])
```

## Train/validation/test split

```python
X = final['tweet']
y = final['winner']

#Train/test split
X_train, X_test, y_train, t_test = train_test_split(X, y, test_size=0.1, shuffle=True, stratify=y)

#Train/val split
X_train, X_val, y_train, y_val = train_test_split(X_train, y_train, test_size=0.2, shuffle=True, stratify=y_train)
```

## Baseline

```python
print(y_val.mode())
print(accuracy_score(y_val, [y_val.mode()]*len(y_val)))

# print(roc_auc_score(y_val, [y_val.mode()]*len(y_val)))
```

## Naive Bayes

```python
counter = CountVectorizer(
    stop_words='english',
    tokenizer=tokenize,
    ngram_range=(1,3),
    analyzer='word'
)

bayes = MultinomialNB()

pipeline = Pipeline([('counter', counter),
                     ('bayes', bayes)])

pipeline.fit(X_train, y_train)

pipeline.score(X_val, y_val)
```

```python
eli5.show_weights(bayes, vec=counter, top=10,
                  target_names=y.unique())
```

## Log Regression

```python
counter = CountVectorizer(
    stop_words='english',
    tokenizer=tokenize,
    ngram_range=(1,5),
    analyzer='word'
)

log = LogisticRegression()

pipeline = Pipeline([('counter', counter),
                     ('bayes', log)])

pipeline.fit(X_train, y_train)

pipeline.score(X_val, y_val)
```

```python
eli5.show_weights(log, vec=counter, top=50,
                  target_names=y.unique())
```
